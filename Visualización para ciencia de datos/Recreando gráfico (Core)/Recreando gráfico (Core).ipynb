# Código del Gráfico (Con Partes Eliminadas)

##### El siguiente código tiene partes eliminadas. El estudiante debe completar el código para reproducir el gráfico mostrado en la imagen.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Cargo el dataset
df = sns.load_dataset('tips')

# creo la figura 
fig, ax = plt.subplots(figsize=(12, 6))

# añado las atribuciones al grafico de dispersion
sns.scatterplot(data=df, x='total_bill', y='tip', 
                hue='time', 
                style='smoker', 
                s=80, 
                ax=ax)

ax.set_title('Relación entre Total Bill y Tip con Anotaciones y Flechas', fontsize=20)
ax.set_xlabel('Total Bill ($)', fontsize=15)
ax.set_ylabel('Tip ($)', fontsize=15)
legend = ax.legend(loc='upper right', fontsize='large', shadow=True, title='Detalles')
legend.get_frame().set_alpha(0.1)        # Transparencia


# Añadir cuadrícula
ax.grid(True, which='major', linestyle='--', linewidth=0.5, color='lightgray')

# Añadir anotacion
ax.annotate('Mayor tip en la cena', xy=(50, 10), xytext=(32, 9.5),arrowprops=dict(facecolor='black', shrink=0.05))

# Añadir flecha verde
ax.arrow(35, 5, 7, 3, head_width=0.8, head_length=0.5, fc='green', ec='green')

# Mostrar el gráfico
plt.show()
```
