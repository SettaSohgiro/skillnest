```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO

# Datos de ejemplo en formato CSV
data = """
fecha,ventas
2023-01-01,200
2023-01-02,220
2023-01-03,210
2023-01-04,205
2023-01-05,215
2023-01-06,225
"""

# Cargar los datos en un DataFrame de Pandas
# df = pd.read_csv(pd.compat.StringIO(data)) error en la carga del data por libreria 
df = pd.read_csv(StringIO(data)) # solucion : se importa libreria StringIO 
display("df : ",df)
```

```python

# Convertir la columna 'fecha' a formato datetime

# df['fecha'] = pd.to_datetime(df['fechas']) # al modificar fecha el llamado es a fechas y es aqui es donde da el error
df['fecha'] = pd.to_datetime(df['fecha']) # Corregido

# # Gráfico de líneas de las ventas
# plt.figure(figsize=(10, 6))
# plt.plot(df['fecha'], df['ventas'])
# plt.title("Ventas Diarias")
# plt.xlabel("Fecha")
# plt.ylabel("Ventas")
# plt.show()

# Gráfico de líneas Mejorado
plt.figure(figsize=(10, 6))
plt.plot(df['fecha'], df['ventas'], marker='o', linestyle='-', color='blue')
plt.title("Ventas Diarias - Gráfico de Líneas")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# # Crear un gráfico de dispersión
# plt.scatter(df['fecha'], df['ventas'])
# plt.title("Ventas Diarias")
# plt.xlabel("Fecha")
# plt.ylabel("Ventas")
# plt.show()

# Gráfico de dispersión Mejorado
plt.figure(figsize=(10, 6))
plt.scatter(df['fecha'], df['ventas'], color='red', s=100)
plt.title("Ventas Diarias - Gráfico de Dispersión")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
